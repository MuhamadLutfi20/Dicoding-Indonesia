```python
# Import seaborn and matplotlib with matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python

# Membaca file CSV
data = pd.read_csv("Pokemon.csv")

# Memisahkan data legendaris dan bukan legendaris
legendary_df = data[data['Legendary'] == True]
not_legendary_df = data[data['Legendary'] == False]

# Menentukan atribut yang akan diplot
attributes = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Tentukan warna dan nama kategori
colors = ['#E69F00', '#56B4E9']
names = ['Legendaris', 'Bukan Legendaris']

# Membuat subplot untuk setiap atribut
fig, axs = plt.subplots(nrows=len(attributes), ncols=1, figsize=(8, 10))

# Loop untuk setiap atribut
for i, attr in enumerate(attributes):
    # Ambil data untuk legendaris dan bukan legendaris
    x_legendary = list(legendary_df[attr])
    x_not_legendary = list(not_legendary_df[attr])

    # Membuat histogram untuk setiap atribut
    axs[i].hist([x_legendary, x_not_legendary], bins=12, density=True, color=colors, label=names)
    axs[i].legend()
    axs[i].set_xlabel(attr)
    axs[i].set_ylabel('Niai Dinormalisasi')

# Menyusun tata letak subplot
plt.tight_layout()

# Menampilkan plot histogram
plt.show()
```
